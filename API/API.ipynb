### Explanation of the data :  In this assignment, we will examine the relative popularity of each tool by reviewing the number of related questions posted in Stack Exchange Q & A. In particular, we use the Stack Exchange API (linking to an external site) to get the number of questions with the "Panda" tag and the number of questions with the "R" tag for the three Q & A subcommunities: Stack Overflow (stackoverflow), Cross Validated (stats.stackexchange), and Data Science (datascience.stackexchange)

### Data wrangling steps with explanation 

#### Import package : StackAPI was not installed, so I downloaded the package using terminal and typed "pip install StackAPI". 

```python
import pandas as pd
import matplotlib.pyplot as plt
from stackapi import StackAPI
```

#### You need to use StackAPI to get access to the stackoverflow website, and request infromation about tag using fetch and tags method.

```python
SITE = StackAPI('stackoverflow')
comments = SITE.fetch('tags')
```

```python
comments
```

#### Since the result is in dictionary, I used "for" iteration to find the location of "R" and "pandas"  

```python
for i in  range(0, len(comments["items"])):
    if comments["items"][i]['name'] == "r":
        print(i)
    if comments["items"][i]['name'] == "pandas":
        print(i)
```

#### putting results in the dataframe 

```python
stackoverflow_r = ["stackoverflow", comments["items"][14]['name'], comments["items"][14]["count"]]
stackoverflow_pandas = ["stackoverflow", comments["items"][46]['name'], comments["items"][46]["count"]]
stackoverflow = pd.DataFrame([stackoverflow_r, stackoverflow_pandas],  columns = ['Site','Program Language', 'tag count'])
stackoverflow
```

#### Do same thing for Cross Validated website

```python
SITE = StackAPI('stats')
comments = SITE.fetch('tags')
comments
```

```python
for i in  range(0, len(comments["items"])):
    if comments["items"][i]['name'] == "r":
        print(i)
    if comments["items"][i]['name'] == "pandas":
        print(i)
```

```python
stats_r = [ "Cross Validated", comments["items"][0]['name'], comments["items"][0]["count"]]
stats = pd.DataFrame([stats_r],  columns = ['Site','Program Language', 'tag count'])
stats
```

#### Do same thing for Data Science website

```python
SITE = StackAPI('datascience')
comments = SITE.fetch('tags')
comments
```

```python
for i in  range(0, len(comments["items"])):
    if comments["items"][i]['name'] == "r":
        print(i)
    if comments["items"][i]['name'] == "pandas":
        print(i)
```

```python
datascience_r = ["Data Science", comments["items"][9]['name'], comments["items"][9]["count"]]
datascience_pandas = ["Data Science", comments["items"][17]['name'], comments["items"][17]["count"]]
datascience = pd.DataFrame([datascience_r, datascience_pandas],  columns = ['Site','Program Language', 'tag count'])
datascience
```

## Final calculations and visualizations

```python
frames = [stackoverflow, stats, datascience]
Stack_Exchange_API = pd.concat(frames)
Stack_Exchange_API["Site & Program Language"] = Stack_Exchange_API["Site"] +' - ' + Stack_Exchange_API["Program Language"]
Stack_Exchange_API
```

```python
ax = Stack_Exchange_API.plot.bar( x='Site & Program Language',y="tag count", figsize=(20,10), color = 'silver', title = "The number of tags for R and Pandas")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.show()
```
