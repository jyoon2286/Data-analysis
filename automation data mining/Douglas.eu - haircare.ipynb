```python
url = "https://www.douglas.eu/c/hair/"



def getNavURLs(url):
    subcat = []
    navbar = []
    data = requests.get(url).text
    soup = BeautifulSoup(data, 'html.parser')
    urls = soup.find_all('a',{'class': 'navigation--link'})
    for url in urls:
        navbar.append(url['href'])
        subcat.append(url.text)
        subcat = [elem.strip() for elem in subcat]
    navbar.pop(0)
    subcat.pop(0)
    return list(zip(subcat,navbar))

nav = getNavURLs(url)
```

```python
for subcat, url in nav:
    print('Sub_category is '+ subcat + ' URL is '+ url)
```

```python
def getBrand(url):
    brandname = []
    data = requests.get(url).text
    soup = BeautifulSoup(data, 'html.parser')
    urls = soup.find_all('a',{'class': 'product--manufacturer'})
    for url in urls:
        brandname.append(url.text)
        brandname = [elem.strip() for elem in brandname]
    return brandname
brand = getBrand(url)
```

```python
def getProduct(url):
    linkProduct = []
    data = requests.get(url).text
    soup = BeautifulSoup(data, 'html.parser')
    urls = soup.find_all('a',{'class': 'product--title'})
    for url in urls:
        linkProduct.append(url['href'])
    return linkProduct
productlink = getProduct(url)
```
